Find the lyrics by the artist The Beatles

```python
beatles = data[data.artist=='The Beatles']
beatles.head()
```

Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub('\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer
          
cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Set display options so that the lyrics aren't cut off

```python
pd.set_option('display.max_colwidth', None)
```

```python
beatles[beatles.song=='Imagine'].text
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
          
cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
          
cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

#### Supplementary Activity

* Use a dataset that consists of text document
* Remove special characters
* Compare the similar document using Count Vector and TF-IDF Vectorizer
* Use the cosine similarity to measure the similarity of text

```python
from pathlib import Path

import kagglehub
import pandas as pd

# Download the BBC news dataset from Kaggle.
path = Path(kagglehub.dataset_download("gpreda/bbc-news"))
print("Path to dataset files:", path)

csv_files = sorted(path.rglob("*.csv"))
print("CSV files:", [file.name for file in csv_files])

if not csv_files:
    raise FileNotFoundError("No CSV file was found in the downloaded dataset.")

bbc_data = pd.read_csv(csv_files[0])
bbc_data.head()
```

```python
candidate_text_columns = ["content", "text", "description", "title", "summary"]
text_columns = [column for column in candidate_text_columns if column in bbc_data.columns]

if not text_columns:
    text_columns = list(bbc_data.select_dtypes(include="object").columns)

if not text_columns:
    raise ValueError("The dataset does not contain any text columns.")

label_column = "title" if "title" in bbc_data.columns else text_columns[0]

print("Text columns used:", text_columns)
print("Label column:", label_column)
```

```python
supplementary = bbc_data.copy()
supplementary["document"] = (
    supplementary[text_columns]
    .fillna("")
    .astype(str)
    .agg(" ".join, axis=1)
    .str.replace(r"[^A-Za-z\s]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.lower()
)

supplementary = supplementary[supplementary["document"].ne("")].copy()
supplementary = supplementary.drop_duplicates(subset="document").reset_index(drop=True)
supplementary = supplementary.head(300).copy()
supplementary["label"] = [
    f"Document {index}: {title}"
    for index, title in enumerate(supplementary[label_column].fillna("Untitled document").astype(str), start=1)
    ]

print("Documents analyzed:", len(supplementary))
supplementary[[label_column, "document"]].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(supplementary["document"])

print("CountVectorizer matrix shape:", count_matrix.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf_vectorizer.fit_transform(supplementary["document"])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

count_similarity = cosine_similarity(count_matrix)
tfidf_similarity = cosine_similarity(tfidf_matrix)

def most_similar_pair(similarity_matrix, labels):
    similarity_matrix = similarity_matrix.copy()
    np.fill_diagonal(similarity_matrix, -1)
    row_index, column_index = np.unravel_index(np.argmax(similarity_matrix), similarity_matrix.shape)
    return {
        "document_1": labels.iloc[row_index],
        "document_2": labels.iloc[column_index],
        "cosine_similarity": similarity_matrix[row_index, column_index],
    }

count_result = most_similar_pair(count_similarity, supplementary["label"])

tfidf_result = most_similar_pair(tfidf_similarity, supplementary["label"])
```

```python
results_df = pd.DataFrame([
    {"vectorizer": "CountVectorizer", **count_result},
    {"vectorizer": "TF-IDF", **tfidf_result},
])

results_df
```

#### Conclusion

This activity showed that text similarity can be measured by cleaning text, transforming it into numeric vectors, and comparing those vectors with cosine similarity. It also showed that TF-IDF often gives more meaningful similarity results than CountVectorizer because it highlights important words and downplays common ones.
